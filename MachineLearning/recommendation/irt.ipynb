```python
import pandas as pd
import numpy as np

# 模拟数据
n_students = 10000  # 学生数量
n_items = 1000      # 知识点数量
n_responses = 100000 # 响应总数

np.random.seed(0)
data = {
    "user_id": np.random.choice(range(n_students), n_responses),
    "item_id": np.random.choice(range(n_items), n_responses),
    "response": np.random.randint(0, 2, n_responses)  # 0或1表示错误或正确
}

df = pd.DataFrame(data)
```

```python
import statsmodels.api as sm

# 为每个知识点单独进行逻辑回归
difficulty_params = {}
for item in df['item_id'].unique():
    item_data = df[df['item_id'] == item]
    model = sm.Logit(item_data['response'], sm.add_constant(item_data['user_id']))
    results = model.fit(disp=0)  # disp=0 防止输出过程信息
    # 使用标签索引来获取难度参数
    difficulty_param = results.params.loc['user_id'] if 'user_id' in results.params else np.nan
    difficulty_params[item] = difficulty_param

# 查看一些难度参数
list(difficulty_params.items())[:5]
```
