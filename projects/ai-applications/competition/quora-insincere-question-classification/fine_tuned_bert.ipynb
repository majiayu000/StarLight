```python
train_df, remaining = train_test_split(df, random_state=42, train_size=0.0075, stratify=df.target.values)

# Splitting the remaining dataset into validation and test sets (90% and 10% respectively)
valid_df, _ = train_test_split(remaining, random_state=42, train_size=0.00075, stratify=remaining.target.values)

# Separating majority and minority classes in the training dataset
majority_class = train_df[train_df.target == 0]
minority_class = train_df[train_df.target == 1]

# Downsampling the majority class to match the size of the minority class
downsampled_majority = resample(majority_class,
                                replace=False,  # sample without replacement
                                n_samples=len(minority_class),  # match minority class
                                random_state=42)  # for reproducible results

# Combining the minority class with the downsampled majority class
downsampled_train_df = pd.concat([downsampled_majority, minority_class])

# Shuffle the downsampled training dataset
downsampled_train_df = downsampled_train_df.sample(frac=1, random_state=42)

# Display the shapes of the downsampled training and validation datasets
downsampled_train_df.shape, valid_df.shape
```

```python
downsampled_train_df.target.plot(kind='hist', title = 'Target Distribution');
```

```python
train_df = downsampled_train_df
```

```python
X_train = train_df["question_text"]
y_train = train_df["target"]

X_valid = valid_df["question_text"]
y_valid = valid_df["target"]
```
